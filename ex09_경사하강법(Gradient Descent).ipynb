#### 1. 목표
- 경사 하강법을 파이썬 프로그램으로 만들어보자.

```python
import numpy as np
```

#### 사용할데이터 셋 생성

##### 문제 데이터 생성(공부한 시간)

```python
# 0.0~1.0 사이의 균등분포를 가진 100개의 데이터 랜덤하게 추출
X =np.random.uniform(low=0.0, high=1.0, size=100)
print(X.shape)
print(X)
```

##### 정답 데이터 생성(성적)

```python
y = 10 * X + 5
print(y.shape)
print(y)
```

##### 경사하강법을 이용한 학습

```python
learning_rate = 0.1 #학습률
num_epoch = 800 # 경사하강법을 적용할 최대 횟수

# 초기 가중치와 절편을 셋팅(랜덤)
w =np.random.uniform(low=0.0, high=1.0)
b =np.random.uniform(low=0.0, high=1.0)

print('초기 가중치:', w)
print('초기 절편:', b)
```

```python
for epoch in range(num_epoch):
    #MSE 구하기
    pre_y = w * X + b
    mse =((pre_y-y) **2).mean()
    print("MSE:", mse)
    
    if mse < 0.001 :
        break
        
    # optimizer(GD 알고리즘) 적용
    w = w -learning_rate * ((pre_y-y)*X).mean()
    b = b -learning_rate * ((pre_y-y)*1).mean()
    
    # 중간결과 출력
    print('epoch:',(epoch+1), 'w:',w,'b:',b)
```
